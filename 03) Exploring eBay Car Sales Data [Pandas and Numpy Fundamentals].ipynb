## Introduction

In this project we are tasked with providing insights on the used car market in Germany.

The dataset we will use is a web scrape of eBay Kleinanzeigen, a classifieds section of the German eBay website.

A copy of the dataset can be found [here](https://data.world/data-society/used-cars-data). This analysis, however, will be completed using a 50,000 row sample with some modifications.

### Summary of Results
1) \\$2,000 to $8,000 USD appears to be the sweet spot for the used car price point range.  
2) Vehicle types can be classified into three price groups: high(\~\\$10,000), medium(\~\\$5,000) and low(\~\\$3,000).
 * On average we see that vehicle body types fall into the following pricing categories: 
  * High Price Point: SUV, coupe, and convertible  
  * Medium Price Point: Van, sedan, and crossover  
  * Low Price Point: Compact and 'other'  
 * On average we see that the top five car brands fall into the below pricing categories:  
  * High Price Point: BMW, Mercedez Benz, and Audi  
  * Medium Price Point: Volkswagen  
  * Low Price Point: Opel     
  
3) The top 5 most frequently sold brands are all German car brands, which, considering the source of the dataset, is as we might expect.  
4) People most frequently sell their cars when their odometer is at 50-90k miles.  
 * There doesn't appear to be a notable odometer variance among brands.  
 * There isn't a notable pattern between vehicle type and mean odometer.  

5) There are three car types that are most prevelant in the market:
* compact ~24% of market
* crossocer ~20% of market
* sedan ~29% of market

6) The average length of time before a car is (assumed) sold is 9 days. Unsold cars have an average advertised length of 15 days. 
 * The type of car does not correlate with the length of time a car is in market.

#### Addressing `price` Outliers

During this process we learned there are pricing outliers that will need to be cleaned. We will now take steps to clean these columns.

```python
# create boolean object that will filter pricing to only very high car prices
expensive_makes_and_models_bool = autos['price'] >= 150000

# filter dataframe using boolean object and keep only columns that add useful context
expensive_makes_and_models = autos.loc[expensive_makes_and_models_bool,['price', 'model', 'brand','registration_year']].value_counts().sort_index(ascending=False).head(15)
print(expensive_makes_and_models)
```

Looking at the above and completing a little research on cars, we can see that cars above \\$350,000 start to reach unreasonable prices for the model and brand of the cars for sale.

Because there are so few occurences of cars with unreasonably high prices it will be fine to simply remove them.

One other action we will take is to remove unreasonably low prices from consideration. This requires more discretion than unreasonably high prices. Because ebay is a bidding site, people can post their cars for $0 and allow users to bid for the vehicle. So when we see a price on a vehicle for \\$0 it does not mean that the car will be sold for that price. Rather, the seller may expect people to submit their best offers or place bids that gradually increase the price. For this reason we will assume that cars below \\$100 were just posted and have few or no bids on them yet. 

Let's remove cars priced below \\$100 and above \\$350,000 with the assumption that we are getting vehicles that are closer to their actual sale price.

```python
# remove values below 99 and above 900,000
autos = autos[autos["price"].between(99,900000)] 

# create boolean object that will filter pricing to only very high car prices
expensive_makes_and_models_bool = autos['price'] >= 150000

# filter dataframe using boolean object and keep only columns that add useful context
expensive_makes_and_models = autos.loc[expensive_makes_and_models_bool,['price', 'model', 'brand','registration_year']].value_counts().sort_index(ascending=False).head(15)
print(expensive_makes_and_models)
```

Our dataset now does not include car priced above \\$350,000 or below \\$100. 

The rationale behind this being that we are not too interested in outliers. There are plenty of ads for $0, and that can be considered an ad with an 'unreal' offer. Cars can sell for cheap, however, and the lower limit was set to 100, which is where distribution starts to pick up. 

A similar rationale can be taken for the higher priced cars. Cars can get quite expensive, but there is a point where that becomes unreasonable. After \\$350,000 price shoots up fast and the model and brand don't justify the price. \\$900,000 was used because that was the next highest value after the \\$350,000 porsche 911, which was a brand new luxury sports car at the time this data was scraped (2016).

- The `registration_month` has a numerical month and has 0 for its minimum and 12 for its maximum. This would leave us with 13 months. We should explore whether there are some miscategorized zero values that should begin with 1.

Given the fairly even distribution here, we can't assume that those values categorized as '0' should have been in '1' or January. One hypothesis is that '0' months are uncategorized or null. 

```python
# create a frequency table of the registration month column - here we converted it to percentages using 'normalize'
autos['registration_month'].value_counts(normalize=True, dropna = False).sort_index()
```

To clean the data we will categorize 0 as Nan and all other values equal to their descriptive numerical counterparts. e.g. 1 = January and 12 = December

```python
#Create a dictionary with a key value relationship of month number and descriptive month
month_dic = {1: 'January', 2: 'February', 3:'March', 4: 'April', 
             5: 'May', 6: 'June', 7: 'July', 8: 'August', 
             9: 'September', 10: 'October', 11: 'November', 12: 'December', 0: np.nan}

# use the dictionary to map the descriptive months to the numerical key in the registration_month column
autos['registration_month'] = autos['registration_month'].map(month_dic)

# Check that the values have changed by running the same code as before
autos['registration_month'].value_counts(normalize=True, dropna = False).sort_index()
```

The remapping of our value keys was successful, however they are now out of order. We will recreate that order below.

```python
# Create a list that matches exactly the values we want to sort and in their proper order
months = ["January", 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
# Use pandas categorical to create an ordered relationship for the months list that will be applied to the series
autos['registration_month'] = pd.Categorical(autos['registration_month'], categories = months, ordered = True)
```

```python
# Check that values have changed and drop the NaN values
autos['registration_month'].value_counts(normalize=True, dropna = True).sort_index()
```

`registration_months` tells us the month in which a car was first registered, something that happens when a new car is purchased. We can assume that there is a correlation between `registration_months`and <u>new car</u> purchase behavior. 

There is an uptick in car registrations in the early-mid months of the year, March to July. This may be a seasonal behavior where people have an increased tendency to buy cars in the Spring and Summer months. Insight will be passed to marketing for further investigation.

### Americanizing the Dataset
The data was pulled from a German website and is a sample of the used German car market. However, our audience is assumed to be American, and so data should be changed to make it more for the American reader.

#### Converting `power_ps` to Horsepower
We'll start by cleaning the column of bad data and changing the __power_ps__ column to horsepower.

`power_ps` is the European equivalent of horsepower. 1 PS or metric horsepower(EU) is equivalent to .989632 HP or mechanical horsepower(US). For reference 1,000 HP is attainable, but very rare (and expensive). The conversion formula we will use is HP = (PS / 1.104).

```python
# View dataset in PS
autos['power_ps'][:10]
```

```python
# all rows from PS to HP
autos['power_ps'] = autos['power_ps'] / 1.104
# round answers for readability (assuming we won't be doing scientific analysis)
autos['power_ps'] = autos['power_ps'].round(2)

# view dataset in HP
autos['power_ps'][:10]
```

```python
# rename columns to horse_power
autos.rename({'power_ps':'horse_power'}, axis = 1, inplace = True)

# show changes
autos[:1]
```

We've now converted the values into HP and renamed them. 

Now let's examine the unrealistic HP values.

```python
# boolean that targets HP above 1000
high_hp_bool = autos['horse_power'] > 1000

# create an dataframe object that 
high_hp = autos.loc[high_hp_bool,['horse_power','model','brand','price']].value_counts().sort_index().head(10)
print(high_hp)
```

#### Converting km to mph

The __odometer_km__ values are in ranges, and not exact value form. Therefore, to keep this clean formatting there will be an expense of not having exact 1:1 match in our column conversions. To translate kilometer to mph, we divide km by 1.609. This, not giving divisible by 100 values, will require rounding up our mph values to the nearest thousandth.

```python
# Showing our values in sorted form
print(autos['odometer_km'].sort_values().unique())
```

```python
#create an empty list for our converted values
odom_vals_mph = []

# assign the unique valeus to an object
odom_vals = autos['odometer_km'].sort_values().unique()

# for loop that converts km to mph and adds it to our list
for km_val in odom_vals:
    mph = km_val / 1.609 # conversion function
    mph = int(round(mph, -3)) # round to nearest 1000
    odom_vals_mph.append(mph) # add to dataset
    
print(odom_vals_mph)
```

We have a list of conversion values, now let's create a dictionary that we will use to replace the values in our series.

```python
# empty dictionary where our final key: values will live
km_mph_dic = {} 

# copy of our mph values, which will be slowly diminished to nothing  
cp_mph_vals = odom_vals_mph

# loop we will use to accomplish associating the key: value pairs
for key in odom_vals:
    for value in cp_mph_vals: # nested for loop method
        km_mph_dic[key] = value # creation of the key: value relationship in our dictionary
        cp_mph_vals.remove(value) # removing the used value to prevent it's use in the next iteration
        break # breaking the current loop so that we will continue to the next without continuing to write in new values
        
        
print(km_mph_dic)
```

Next we will use our dictionary to remap our __odomoter_km__ values.

```python
# map our mph values over the mph values for the odometer_km values in the series
autos['odometer_km'] = autos['odometer_km'].map(km_mph_dic)

#check the work
print(autos['odometer_km'].sort_values().unique())
```

### Estimating Car Sales
One important metric, which we don't have access to, is knowing which cars sold and at what price. We can try to estimate this by using the `ad_created` and `last_seen` columns, along with the context that the last day that the website was crawled is April 7th. We will assume that if a last_seen date is before April 7th that the car has been sold at the price on that date. The `ad_created` column will let us for how long and ad was created before the ad was taken down and assumed sold. 

Let's start by splitting the dataset into sold and not sold.
#### Splitting the Dataset

```python
cutoff = autos['ad_created'].max() # April 7th 2016
print(cutoff)

sold_autos = autos[autos["last_seen"] < cutoff]
unsold_autos = autos[autos["last_seen"] >= cutoff]

print("Sold During Scrape Period:", len(sold_autos))
print("Unsold During Scrape Period:", len(unsold_autos))
```

Let's also remove hours minutes and seconds from `last_seen` so that we remove unnecessary noise.

```python
# copy inplace of itself to prevent errors
sold_autos = sold_autos.copy()
unsold_autos = unsold_autos.copy()

# function to remove datetime HMS
def no_hms(datetime_object):
    new_datetime_object = datetime_object.replace(hour = 0, minute = 0, second = 0)
    return new_datetime_object

# apply to new column
sold_autos['last_seen_day'] = sold_autos['last_seen'].apply(no_hms)
unsold_autos['last_seen_day'] = unsold_autos['last_seen'].apply(no_hms)
```

#### Time Before Sale
Let's calculate the average length of time an ad is up before it sold.

```python
# add new column
sold_autos["ad_lifetime"] = sold_autos["last_seen_day"] - sold_autos["ad_created"] + dt.timedelta(days = 1) # we add one day for smoothing
unsold_autos["ad_lifetime"] = unsold_autos["last_seen_day"] - unsold_autos["ad_created"] + dt.timedelta(days = 1) # we add one day for smoothing
```
