# Storytelling Data Visualization on Exchange Rates

This project will be an exercise in explanatory data visualizations. 

The subject will be a dataset of [Euro daily exchange rates](https://www.kaggle.com/lsind18/euro-exchange-daily-rates-19992020).

The goal of our analysis is to understand how exchange rates between the Euro and other world currencies changed over time. There will be a focus on the US Dollar to Euro exchange rate.

## The Exchange Rate Dataset

Daria Chemkaeva put together the data set and made it available on [Kaggle](https://www.kaggle.com/lsind18/euro-exchange-daily-rates-19992020).

```python
import pandas as pd
exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')
```

The dataset contains 5,699 rows and 41 columns of country exchange rate data. There are several columns with null values present, and there is a mix of object and float types. Each row is a daily snapshot of the exchange rates for countries' currencies compared to the Euro. 

We read exchange rates by understanding the following: If the exchange rate of the euro to the US dollar is 1.5, you get 1.5 US dollars if you pay 1.0 euro (one euro has more value than one US dollar at this exchange rate).

# Data Cleaning

```python
# rename the country columns to be more readable
#create a variable of just country column names
country_column_names = exchange_rates.columns[1:]
# remove unecessary characters
for c_name in country_column_names:
    c_n_name = c_name
    c_n_name = c_n_name.replace('[','')
    c_n_name = c_n_name.replace(' ]', '')
    c_n_name = c_n_name.replace(' ', '_')
    #replace the old column name with the new
    exchange_rates.rename(columns = {c_name:c_n_name}, inplace = True)
```

```python
# rename the Period\Unit to Time
exchange_rates.rename(columns={'Period\\Unit:':'Time'}, inplace = True) # we use two backslashes as an escape
```

```python
print(exchange_rates.columns)
```

```python
# transform the time column to datetime
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
```

```python
# sort the columns by time
exchange_rates.sort_values('Time', inplace=True)
```

```python
# reset the index to follow the new sort
exchange_rates.reset_index(drop=True, inplace=True)
```

```python
exchange_rates
```

```python
# copy the time and us dollar to a variable
euro_to_dollar = exchange_rates[['Time','US_dollar']].copy()
```

```python
# check for data integrity of US dollar column
euro_to_dollar['US_dollar'].value_counts()
```

```python
# remove rows with '-' character
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar']!= '-']
```

```python
# convert to float type
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
```

```python
euro_to_dollar.info()
```

## Rolling Mean

```python
import matplotlib.pyplot as plt
# Enables Jupyter to display graphs
%matplotlib inline 
```

```python
plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
plt.show()
```

```python
# create a new series that shows the rolling mean over 30 days of the USD-Euro exchange rate
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
```

```python
plt.plot(euro_to_dollar['Time'], euro_to_dollar['rolling_mean'])
plt.show()
```

## Parlimentary Elections in the EU

```python
# average rate for exchange rate for the whole time period
total_mean = round(euro_to_dollar['US_dollar'].mean(),2)
print(total_mean)
```

```python
# European Parliamentary Elections datetime timelines
EU_election_1 = euro_to_dollar.copy()[euro_to_dollar['Time'] < '2004-7-20']
EU_election_2 = euro_to_dollar.copy()[(euro_to_dollar['Time'] >= '2004-7-20') & (euro_to_dollar['Time'] < '2009-7-20')]
EU_election_3 = euro_to_dollar.copy()[(euro_to_dollar['Time'] >= '2009-7-14') & (euro_to_dollar['Time'] < '2014-7-1')]
EU_election_4 = euro_to_dollar.copy()[(euro_to_dollar['Time'] >= '2014-7-2') & (euro_to_dollar['Time'] < '2019-7-2')]
EU_election_5 = euro_to_dollar.copy()[(euro_to_dollar['Time'] >= '2019-7-2')]
```

```python
#list of years where there is an election
election_years = ['1999', '2004', '2009', '2014', '2019']
```

```python
# mean of the years where an elected parliment is in power
E_1_mean = round(EU_election_1['US_dollar'].mean(),2)
E_2_mean = round(EU_election_2['US_dollar'].mean(),2) 
E_3_mean = round(EU_election_3['US_dollar'].mean(),2)
E_4_mean = round(EU_election_4['US_dollar'].mean(),2)
E_5_mean = round(EU_election_5['US_dollar'].mean(),2)
```

```python
# add the FiveThirtyEight style
import matplotlib.style as style
style.use('fivethirtyeight')

# add the plots
fig, (ax1, ax2, ax3, ax4, ax5) = plt.subplots(nrows = 5, ncols = 1, figsize = (6,8))
axes = [ax1, ax2, ax3, ax4, ax5]
# add the base data to the plots and change base formatting
for ax in axes:
    ax.plot(euro_to_dollar['Time'], euro_to_dollar['rolling_mean'],color ='#003399', linewidth=2.5)
    ax.set_yticklabels([])
    ax.set_xticklabels([])
    ax.grid(axis = 'x', b=None)
    ax.grid(axis = 'y', alpha = 0.25)

# overlay the highlighted time periods
ax1.plot(EU_election_1['Time'], EU_election_1['rolling_mean'], color = '#FFCC00')
#add y-axis labels
ax1.text(9500, 1.45,'$1.5', alpha = .5)
ax1.text(9500, 0.95,'$1.0', alpha = .5)
#add titles and subtitles
ax1.text(9500,1.8,'The mean Euro-USD exchange is $1.2 USD', weight = 'bold')
ax1.text(9500,1.65,'Euro to USD exchange rates between EU parliamentary elections', size = 12)
# add the mean for the highlighted period
ax1.text(11100, 1.20, s = '$'+str(E_1_mean), alpha = .5)

# x-axis labels of election years
for i, year in zip(range(6), election_years):
    ax1.text(x = 10000 + (i*2000), y = 0.7, s = year, alpha = .5, size = 10)
# add the rest of the highlighted overlays and the mean for the period
ax2.plot(EU_election_2['Time'], EU_election_2['rolling_mean'], color = '#FFCC00')
ax2.text(13100, 1.05, s = '$'+str(E_2_mean), alpha = .5)
ax3.plot(EU_election_3['Time'], EU_election_3['rolling_mean'],color = '#FFCC00')
ax3.text(14900, 1.05, s = '$'+str(E_3_mean),alpha = .5)
ax4.plot(EU_election_4['Time'], EU_election_4['rolling_mean'], color = '#FFCC00')
ax4.text(16800, 1.35, s = '$'+str(E_4_mean), alpha = .5)
ax5.plot(EU_election_5['Time'], EU_election_5['rolling_mean'], color = '#FFCC00')
ax5.text(18200, 1.35, s = '$'+str(E_5_mean), alpha = .5)
ax5.text(9450, 0.5,'Kevin Darley'+ ' '*40 +'Source: European Central Bank', color = '#f0f0f0', backgroundcolor = '#4d4d4d', size = 12)

plt.show()
```
